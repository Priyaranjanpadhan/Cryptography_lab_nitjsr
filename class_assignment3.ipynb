```python
text = input("The encrypted text is : ").lower()

def inverse_key(multiplicative_key):
    for i in range(1, 26):
        if(((multiplicative_key * i) % 26) == 1):
            return i
    return None

def plain_text(multiplicative_key):
    cipher = []
    if(inverse_key(multiplicative_key) is None):
        return None
    else:
        for ch in text:
            cipher.append(chr((((ord(ch) - ord('a')) * inverse_key(multiplicative_key)) % 26) + ord('a')))
    return cipher

for i in range(0,26):
    value = plain_text(i)
    if(value != None):
        print("the plain text for the given text is : ", "".join(value), f"and the key is {i}")
```
